```python
import ROOT
from array import array
import numpy as np
```

```python
MyTree = ROOT.TTree( 'tree1', 'tree para 3 datos' )
```

```python
x = array( 'f', [ 0 ] )
y = array( 'f', [ 0 ] )
z = array( 'f', [ 0 ] )
```

```python
MyTree.Branch( 'x', x, 'x/F' )
MyTree.Branch( 'y', y, 'y/F' )
MyTree.Branch( 'z', z, 'z/F' )
```

```python
for i in range(100000):
  x[0] = np.random.randn()*5.+5.
  y[0] = np.random.randn()*5.+5.
  z[0] = np.random.rand()*10.
  MyTree.Fill()
```

```python
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x*y>>h1(10,0,100)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("x*z>>h2(10,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("x*y*z>>h3(10,0,1000)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("x>>h4(10,0,10)","z>=4.5 & z<=5.5")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("x:z>>h5(10,0,10,10,0,10)","y>=4.5 & y<=5.5","lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```python
MyTree.Draw("x:y:z>>h6(10,0,10,10,0,10,10,0,10)","","lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
